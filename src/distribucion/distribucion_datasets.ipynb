```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
```

```python
# Load the CSV file into a DataFrame
nissan = pd.read_csv("nissan-dataset.csv")


# Display the first few rows of the DataFrame
nissan.head()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(nissan['price'], kde=True, color='blue')
plt.title('')
plt.xlabel('Estimación')
plt.ylabel('Frecuencia')
plt.show()

mean = np.mean(nissan['price'])
std = np.std(nissan['price'])

print(f"Mean: {mean}")
print(f"Standard Deviation: {std}")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(nissan['price'], kde=True, stat='density', bins=30, color='blue', alpha=0.6 , label='Data')

# Fit a lognormal distribution to the data
shape, loc, scale = stats.lognorm.fit(nissan['price'], floc=0)

# Overlay Lognormal distribution
x = np.linspace(min(nissan['price']), max(nissan['price']), 100)
pdf_lognormal = stats.lognorm.pdf(x, shape, loc, scale)
plt.plot(x, pdf_lognormal, 'r-', label='Lognormal Distribution')

plt.title('')
plt.xlabel('x')
plt.ylabel('f(x)')


plt.legend()
plt.savefig('goals_distribution.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
