```python
import pandas as pd

# Load the dataset
file_path = 'datasets/Carbon_Emission.csv'
data = pd.read_csv(file_path)

# Display the first few rows and columns information
data.head(), data.info()
```

```python
# Identify columns with lists of strings
list_columns = ['Recycling', 'Cooking_With']

# Process each column
for col in list_columns:
    # Clean the strings in the list and extract unique items
    data[col] = data[col].apply(lambda x: eval(x) if isinstance(x, str) else x)  # Convert string lists to actual lists
    unique_items = set([item for sublist in data[col] for item in sublist])
    
    # Create a new column for each unique item, indicating its presence (1) or absence (0)
    for item in unique_items:
        data[f"{col}_{item}"] = data[col].apply(lambda x: 1 if item in x else 0)

# Drop the original list columns to avoid redundancy
data_transformed = data.drop(columns=list_columns)

# Display the first few rows of the transformed dataset
data_transformed.head()
```

```python
# Save the transformed data to a new CSV file
output_path = 'datasets/Carbon_Emission_transformed.csv'
data_transformed.to_csv(output_path, index=False)

output_path
```

```python
# Load the dataset
file_path = 'datasets/wind_dataset.csv'
data = pd.read_csv(file_path)

# Display the first few rows and columns information
data.head(), data.info()
```

```python
# Separar la columna 'DATE' en día, mes y año
data[['YEAR', 'MONTH', 'DAY']] = data['DATE'].str.split('-', expand=True)

# Eliminar la columna original si ya no la necesitas
df = data.drop(columns=['DATE'])

# Guardar el nuevo archivo con las columnas separadas
df.to_csv('datasets/wind_dataset_transformed.csv', index=False)

df.head()
```
