# Modelo IQR

```python
import numpy as np
import pandas as pd
```

## Dataset Wind

```python
# Replace 'your_file.npy' with the path to your .npy file
file_path = '../resultados_grid_search/Wind/Wind_IQR.npy'
data_wind = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_depth', 'MSE']

# Create a DataFrame
df_wind = pd.DataFrame(data_wind, columns=columns)

# Display the DataFrame
df_wind
```

```python
sorted_df_wind = df_wind.sort_values(by='MSE', ascending=True)
sorted_df_wind['Rank'] = range(1, len(sorted_df_wind) + 1)

top_500_wind = sorted_df_wind[:250]
top_500_wind
```

## Dataset Carbon Emission

```python
file_path = '../resultados_grid_search/Carbon_Emission/Carbon_Emission_IQR.npy'
data_carbon = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_depth', 'MSE']

# Create a DataFrame
df_carbon = pd.DataFrame(data_carbon, columns=columns)

# Display the DataFrame
df_carbon
```

```python
sorted_df_carbon = df_carbon.sort_values(by='MSE', ascending=True)
sorted_df_carbon['Rank'] = range(1, len(sorted_df_carbon) + 1)

top_500_carbon = sorted_df_carbon[:250]
top_500_carbon
```

## House 8L

```python
file_path = '../resultados_grid_search/House_8L/House_8L_IQR.npy'
data_house_8L = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_depth', 'MSE']

# Create a DataFrame
df_house_8L = pd.DataFrame(data_house_8L, columns=columns)

# Display the DataFrame
df_house_8L
```

```python
sorted_df_house_8L = df_house_8L.sort_values(by='MSE', ascending=True)
sorted_df_house_8L['Rank'] = range(1, len(sorted_df_house_8L) + 1)

top_500_house_8L = sorted_df_house_8L[:250]
top_500_house_8L
```

```python
# Unimos df1 y df2 usando las columnas clave col1, col2 y col3
df_merged = pd.merge(top_500_wind, top_500_carbon, on=['n_estimators', 'group_size', 'max_depth'], how='inner')

# Luego, unimos el resultado con df3
df_final = pd.merge(df_merged, top_500_house_8L, on=['n_estimators','group_size', 'max_depth'], how='inner')

# Mostramos el dataframe final
print(df_final)
```
