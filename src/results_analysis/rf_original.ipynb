```python
import numpy as np
import pandas as pd
```

```python
file_path = '../resultados_grid_search/Carbon_Emission/Carbon_Emission_Random_Forest.npy'
data_carbon = np.load(file_path)

# Define column names
columns = ['max_depth', 'MSE']

df_carbon = pd.DataFrame(data_carbon, columns=columns)

sorted_df_carbon = df_carbon.sort_values(by='MSE', ascending=True)

# Add rank column
sorted_df_carbon['rank'] = sorted_df_carbon['MSE'].rank(method='min')

sorted_df_carbon[:5]
```
