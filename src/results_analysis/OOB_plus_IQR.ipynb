# Modelo IQR

```python
import numpy as np
import pandas as pd
```

## Dataset Wind

```python
# Replace 'your_file.npy' with the path to your .npy file
file_path = '../resultados_grid_search/Wind/Wind_OOB_plus_IQR.npy'
data_wind = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_depth', 'MSE']

# Create a DataFrame
df_wind = pd.DataFrame(data_wind, columns=columns)
```

```python
sorted_df_wind = df_wind.sort_values(by='MSE', ascending=True)

top_50_wind = sorted_df_wind[:50]
top_50_wind
```

## Dataset Carbon Emission

```python
file_path = '../resultados_grid_search/Carbon_Emission/Carbon_Emission_OOB_plus_IQR.npy'
data_carbon = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_depth', 'MSE']

# Create a DataFrame
df_carbon = pd.DataFrame(data_carbon, columns=columns)
```

```python
sorted_df_carbon = df_carbon.sort_values(by='MSE', ascending=True)

top_50_carbon = sorted_df_carbon[:50]
top_50_carbon
```

## House 8L

```python
file_path = '../resultados_grid_search/House_8L/House_8L_OOB_plus_IQR.npy'
data_house_8L = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_depth', 'MSE']

# Create a DataFrame
df_house_8L = pd.DataFrame(data_house_8L, columns=columns)
```

```python
sorted_df_house_8L = df_house_8L.sort_values(by='MSE', ascending=True)

top_50_house_8L = sorted_df_house_8L[:50]
top_50_house_8L
```

```python
# Drop the 'MSE' column from each top_50 DataFrame
top_50_wind_dropped = top_50_wind.drop(columns=['MSE'])
top_50_carbon_dropped = top_50_carbon.drop(columns=['MSE'])
top_50_house_8L_dropped = top_50_house_8L.drop(columns=['MSE'])

# Concatenate the DataFrames
concatenated_df = pd.concat([top_50_carbon_dropped, top_50_house_8L_dropped, top_50_wind_dropped], ignore_index=True)

# Drop repeated rows
concatenated_df = concatenated_df.drop_duplicates()

# Convert all values to integers
concatenated_df = concatenated_df.astype(int)

# Save the concatenated DataFrame to a CSV file
concatenated_df.to_csv('csvs/OOB_plus_IQR_Top_150.csv', index=False)
```

# Top 150

```python
file_path = '../resultados_top_150/Wind/Wind_OOB_plus_IQR_150.npy'
data_wind = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_depth', 'MSE']

df_wind_150 = pd.DataFrame(data_wind, columns=columns)

sorted_df_wind_150 = df_wind_150.sort_values(by='MSE', ascending=True)

# Add rank column
sorted_df_wind_150['rank'] = sorted_df_wind_150['MSE'].rank(method='min')

sorted_df_wind_150[:50]
```
