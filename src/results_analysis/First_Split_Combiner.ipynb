# Modelo First Splits Combiner

```python
import numpy as np
import pandas as pd
```

## Dataset Wind

```python
# Replace 'your_file.npy' with the path to your .npy file
file_path = '../resultados_grid_search/Wind/Wind_First_Splits_Combiner.npy'
data_wind = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_features', 'MSE']

# Create a DataFrame
df_wind = pd.DataFrame(data_wind, columns=columns)
```

```python
sorted_df_wind = df_wind.sort_values(by='MSE', ascending=True)

top_50_wind = sorted_df_wind[:50]
top_50_wind
```

## Dataset Carbon Emission

```python
file_path = '../resultados_grid_search/Carbon_Emission/Carbon_Emission_First_Splits_Combiner.npy'
data_carbon = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_features', 'MSE']

# Create a DataFrame
df_carbon = pd.DataFrame(data_carbon, columns=columns)
```

```python
sorted_df_carbon = df_carbon.sort_values(by='MSE', ascending=True)

top_50_carbon = sorted_df_carbon[:50]
top_50_carbon
```

## House 8L

```python
file_path = '../resultados_grid_search/House_8L/House_8L_First_Splits_Combiner.npy'
data_house_8L = np.load(file_path)

# Define column names
columns = ['n_estimators', 'group_size', 'max_features', 'MSE']

# Create a DataFrame
df_house_8L = pd.DataFrame(data_house_8L, columns=columns)
```

```python
sorted_df_house_8L = df_house_8L.sort_values(by='MSE', ascending=True)

top_50_house_8L = sorted_df_house_8L[:50]
top_50_house_8L
```

```python
# Drop the 'MSE' column from each top_50 DataFrame
top_50_wind_dropped = top_50_wind.drop(columns=['MSE'])
top_50_carbon_dropped = top_50_carbon.drop(columns=['MSE'])
top_50_house_8L_dropped = top_50_house_8L.drop(columns=['MSE'])

# Concatenate the DataFrames
concatenated_df = pd.concat([top_50_carbon_dropped, top_50_house_8L_dropped, top_50_wind_dropped], ignore_index=True)

# Drop repeated rows
concatenated_df = concatenated_df.drop_duplicates()

# Convert 'max_features' column to float and ensure negative values are preserved
concatenated_df['max_features'] = concatenated_df['max_features'].astype(float)
concatenated_df['n_estimators'] = concatenated_df['n_estimators'].astype(int)
concatenated_df['group_size'] = concatenated_df['group_size'].astype(int)

# Save the concatenated DataFrame to a CSV file with specified float format for 'max_features'
concatenated_df.to_csv('csvs/First_Splits_Combiner_Top_150.csv', index=False, float_format='%.2f', columns=['n_estimators', 'group_size', 'max_features'])
```
